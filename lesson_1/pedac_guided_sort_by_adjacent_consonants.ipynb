```python
VOWELS = ["a", "e", "i", "o", "u"]



def clean_string(input_str):
    output_str = input_str.lower()
    output_str = output_str.replace(" ", "")

    return output_str

def count_max_adjacent_consonants(input_str):
    max_adj = 0
    loop_max = 0

    for char in input_str:
        if char in VOWELS:
            loop_max = 0
            continue

        loop_max += 1

        if loop_max > max_adj:
            max_adj = loop_max

    if max_adj == 1:
            return 0

    return max_adj

def find_index(input_str, current_list):
    new_adj_cons = count_max_adjacent_consonants(input_str)

    for idx, elem in enumerate(current_list):
        elem = clean_string(elem)
        elem_adj = count_max_adjacent_consonants(elem)


        if new_adj_cons > elem_adj:
            return idx

        if new_adj_cons <= elem_adj:
            if idx == (len(current_list) - 1):
                return (idx + 1)

            if count_max_adjacent_consonants(current_list[idx + 1]) <= elem_adj:
                print("continues")
                continue

            return (idx + 1)


def sort_by_consonant_count(input_list):
    output_list = []

    for elem in input_list:
        elem_str = clean_string(elem)

        if len(output_list) == 0:
            output_list.append(elem)
            continue

        idx_to_add = find_index(elem_str, output_list)

        if idx_to_add == len(output_list):
            output_list.append(elem)
            continue
        
        output_list.insert(idx_to_add, elem)

    return output_list


a_list = ["a"]
my_list = ['can can', 'toucan', 'batman', 'salt pan'] 

print(count_max_adjacent_consonants('can can'))
print(count_max_adjacent_consonants("salt pan"))
print(sort_by_consonant_count(my_list))
```
